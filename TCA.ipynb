# set the SeniorCitizen data 1 and 0 to yes/no to make it easy to understand

```python
# set the SeniorCitizen data 1 and 0 to yes/no to make it easy to understand
def conv(value): 
    if value == 1: 
        return "yes" 
    else: 
        return "no" 
        
df['SeniorCitizen'] = df["SeniorCitizen"].apply(conv)
```

```python

# List of columns to plot
columns = ['PhoneService', 'MultipleLines', 'InternetService',
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
           'TechSupport', 'StreamingTV', 'StreamingMovies']

# Create subplots (3 rows, 3 columns)
fig, axes = plt.subplots(3, 3, figsize=(15, 12))  
fig.suptitle("Count Plots of Various Services (with Churn)", fontsize=16)  

# Loop through each column and create a count plot with hue='Churn'
for ax, col in zip(axes.flat, columns):
    sns.countplot(x=df[col], hue=df['Churn'], ax=ax)  # Added hue='Churn'
    ax.set_title(col)
    ax.bar_label(ax.containers[0])  # Add count labels

# Adjust layout for better spacing
plt.tight_layout(rect=[0, 0, 1, 0.96])  
plt.show()
```

#Customers without support services (online security, backup, tech support) have higher churn.
Fiber optic internet users also show more churn compared to DSL users. Customers without internet services have the lowest churn.
